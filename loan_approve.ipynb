```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('clean.csv')
```

```python
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["loan_approved"].unique()
```

```python
df["loan_approved"].dtype
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df['loan_approved'].unique()
```

```python
loan_approved_dict ={
    False:0,  True:1
}
```

```python
df['loan_approved'] = df['loan_approved'].map(loan_approved_dict)
df
```

```python
x=df.drop('loan_approved',axis=1)
```

```python
y=df[['loan_approved']]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

SUPPRT VECTOR CLASSIFIER

```python
from sklearn.svm import SVC
```

```python
sv=SVC()
sv.fit(x_train,y_train)
```

```python
print('training data',sv.score(x_train,y_train))
print('testing data',sv.score(x_test,y_test))
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
print('training data',lr.score(x_train,y_train))
print('testing data',lr.score(x_test,y_test))
```

DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
import joblib
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
print('training data',dtc.score(x_train,y_train))
print('testing data',dtc.score(x_test,y_test))
```

```python
joblib.dump(dtc,'dtc_load.lb')
```

```python
model=joblib.load('dtc_load.lb')
```

```python
pred =dtc.predict(x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,classification_report
print(confusion_matrix(pred,y_test))
print(accuracy_score(pred,y_test))
precision_score(pred,y_test)
```

```python
print(classification_report(pred,y_test))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(model,filled=True)
plt.show()
```
